```python
import pathlib
import shlex
import shutil
import subprocess
```

```python
shutil.rmtree("data/gwas", ignore_errors=True)
gwas_root = pathlib.Path("data/gwas/")
gwas_root.mkdir()
```

```python
pheno_root = pathlib.Path("data/pheno")
pheno_paths = [
    p for p in pheno_root.glob("*.tsv") 
    if not p.stem.startswith("anon_")
]
print([p.stem for p in pheno_paths])
```

```python
for pheno_path in pheno_paths:
    pheno_type = pheno_path.stem
    print(f"Running GWAS for {pheno_type}")

    output_dir = gwas_root.joinpath(pheno_type)
    output_dir.mkdir(exist_ok=True)

    command = f"""
    plink2 \
        --pfile ../../data/geno/ukb_wb_subsampled \
        --pheno {pheno_path.as_posix()} \
        --glm zs hide-covar allow-no-covars \
        --threads 110 \
        --out {output_dir.joinpath('result').as_posix()}
    """
    result = subprocess.run(shlex.split(command))
```
